```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

db = load_wine()
x = db.data
y = db.target

df = pd.DataFrame(data=x, columns=db.feature_names)
df['target'] = y
x = df.drop('target', axis=1)
y = df['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
y_pred = model_rf.predict(x_test)
accuracy = model_rf.score(x_test, y_test)
print(f'Acurácia do modelo RandomForestClassifier: {accuracy:.2f}')

model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)
y_pred = model_knn.predict(x_test)
accuracy = model_knn.score(x_test, y_test)
print(f'Acurácia do modelo KNeighborsClassifier: {accuracy:.2f}')

model_svm = SVC()
model_svm.fit(x_train, y_train)
y_pred = model_svm.predict(x_test)
accuracy = model_svm.score(x_test, y_test)
print(f'Acurácia do modelo SVC: {accuracy:.2f}')
```

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor

db = fetch_california_housing()
x = pd.DataFrame(db.data, columns=db.feature_names)
y = pd.Series(db.target, name='MedHouseVal')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model_lr = LinearRegression()
model_lr.fit(x_train, y_train)
score = model_lr.score(x_test, y_test)
print(f'Acurácia do modelo LinearRegression: {score:.2f}')

model_rg = Ridge()
model_rg.fit(x_train, y_train)
score = model_rg.score(x_test, y_test)
print(f'Acurácia do modelo Ridge: {score:.2f}')

model_rf = RandomForestRegressor()
model_rf.fit(x_train, y_train)
score = model_rf.score(x_test, y_test)
print(f'Acurácia do modelo RandomForestRegressor: {score:.2f}')
```

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

centers = [[1, 1], [-1, -1], [1, -1]]
x, y = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0)

x_scaled = StandardScaler().fit_transform(x)

plt.scatter(x_scaled[:, 0], x_scaled[:, 1])
plt.show()
```

```python
from sklearn.cluster import KMeans, DBSCAN
import seaborn as sns

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(x_scaled)

dbscan = DBSCAN(eps=0.3, min_samples=10)
dbscan_labels = dbscan.fit_predict(x_scaled)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=x_scaled[:, 0], y=x_scaled[:, 1], hue=kmeans_labels, palette='viridis', s=50)
plt.title('K-Means')

plt.subplot(1, 2, 2)
sns.scatterplot(x=x_scaled[:, 0], y=x_scaled[:, 1], hue=dbscan_labels, palette='viridis', s=50)
plt.title('DBSCAN')

plt.show()
```

```python
import yfinance as yf
import pandas as pd
```

```python
arm = yf.download("ARM", start="2024-07-01", end="2024-08-31")
```

```python
arm['Change'] = ((arm['Close'] - arm['Open']) / arm['Open']) * 100
arm['Vol'] = arm['Volume'].apply(lambda x: f"{x:,}")
```
